```python
#Checking if outliers exists or not using code
Q1 = df["Profit"].quantile(0.25)
Q3 = df["Profit"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df[(df["Profit"] < lower_bound) | 
   (df["Profit"] > upper_bound)]
```

```python
df.loc[(df["Profit"] < lower_bound) | 
       (df["Profit"] > upper_bound), "Profit"]#displaying rows with Outliers in profit column
```

```python
len(df.loc[(df["Profit"] < lower_bound) | #Finding numner of outliers present in Profit Column
           (df["Profit"] > upper_bound)])
```

```python
df["Profit"].fillna(df["Profit"].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
#Checking outliers for shipping cost column
Q1 = df["Shipping_Cost"].quantile(0.25)
Q3 = df["Shipping_Cost"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df[(df["Shipping_Cost"] < lower_bound) | 
   (df["Shipping_Cost"] > upper_bound)]
```

```python
len(df.loc[(df["Shipping_Cost"] < lower_bound) | #Finding numner of outliers present in Shipping_cost Column
           (df["Shipping_Cost"] > upper_bound)])
```

```python
df["Shipping_Cost"].fillna(df["Shipping_Cost"].median(),inplace=True)#filling null values with median in shipping cost column
```

```python
df.isna().sum()#Checking null values
```

```python
df.duplicated().sum()#Finding total number of dupliacte values
```

```python
df["Order_ID"].duplicated().sum() #gives total number of duplicates in Order_ID Column
```

```python
df.duplicated().sum()#checking whether duplicates are removed or not
```

```python
df.info()#data types, non-null counts, and overall dataset size.
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"], errors="coerce")
df["Ship_Date"] = pd.to_datetime(df["Ship_Date"], errors="coerce") #Converting date columns data type(to datetime)
```

```python
df.info()
```

```python
#checking how many invalid rows are deleted
before = len(df)
df = df[df["Ship_Date"] >= df["Order_Date"]]
after = len(df)
print("Invalid rows removed:", before - after)
```

```python
df = df[df["Sales"] >= 0] #Removing Negative Sales
```

```python
 df = df[df["Discount"] <= 0.8]#for extreme discount
```

```python
    Q1 = df["Shipping_Cost"].quantile(0.25) 
    Q3 = df["Shipping_Cost"].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df["Shipping_Cost"] = df["Shipping_Cost"].clip(lower, upper)#the values less than lower are replaced with lower values tha values that are higher
                                                                #than the upper values are replaced by upper value 
```

```python
sns.boxplot(y=df["Profit"])
plt.title("Profit Before outlier capping")
plt.show()#Plotting box plot for profit column
```

```python
#Managing outliers in profit column
Q1 = df["Profit"].quantile(0.25)
Q3 = df["Profit"].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
df["Profit"] = df["Profit"].clip(lower, upper)
```

```python
sns.boxplot(y=df["Profit"])
plt.title("Profit After outlier capping")
plt.show()#Plotting box plot for profit column after outliers treatment
```

```python
df["Product_Category"].unique()#checking unique values in product category column
```

```python
df["Product_Category"].value_counts()#Checking frequency using value_counts()
```

```python
#standardizing all categorical columns
df["Product_Category"] = df["Product_Category"].str.strip().str.title()
df["Customer_Segment"] = df["Customer_Segment"].str.strip().str.title()
df["Region"] = df["Region"].str.strip().str.title()
df["State"] = df["State"].str.strip().str.title()
df["Sub_Category"] = df["Sub_Category"].str.strip().str.title()
df["Payment_Mode"] = df["Payment_Mode"].str.strip().str.title()
```

```python
#Top 20% customers revenue contribution (Pareto Analysis)
customer_sales = df.groupby("Customer_ID")["Sales"].sum().sort_values(ascending=False)#Calculate total sales per customer
top20 = int(len(customer_sales)*0.2)#Select top 20% customers
customer_sales.head(top20).sum() / customer_sales.sum()
```

```python
top20_sales = customer_sales.head(top20).sum()
remaining_sales = customer_sales.sum() - top20_sales
plt.figure(figsize=(6,4))
plt.bar(["Top 20% Customers", "Remaining 80% Customers"],
        [top20_sales, remaining_sales])
plt.title("Revenue Contribution Comparison")
plt.ylabel("Total Sales")
plt.show()
```

```python
#Repeat customer behavior
df["Customer_ID"].value_counts()
```

```python
customer_counts = df["Customer_ID"].value_counts()
repeat_customers = (customer_counts > 1).sum()# Customers who ordered more than once
total_customers = df["Customer_ID"].nunique()# Total unique customers
repeat_percentage = repeat_customers / total_customers * 100
repeat_percentage
```

```python
one_time = (customer_counts == 1).sum()
repeat = (customer_counts > 1).sum()
plt.figure(figsize=(6,4))
plt.bar(["One-Time Customers", "Repeat Customers"], [one_time, repeat])
plt.title("Customer Purchase Frequency")
plt.ylabel("Number of Customers")
plt.show()
```

```python
#Visual-16
# Visualizing correlation between numerical variables
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Quantity also shows a moderate positive correlation with Sales (0.35) and Profit (0.32), meaning larger orders tend to generate more revenue and profit. Discount has a slight negative correlation with Sales (-0.088) and Profit (-0.075), indicating that higher discounts can reduce profitability. Shipping Cost shows almost no correlation with Profit, suggesting it does not strongly affect profit levels.
